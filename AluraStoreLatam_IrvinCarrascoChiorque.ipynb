### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Ingreso total por cada tienda (creación de variables y ingreso total)
t1 = tienda["Precio"].sum()
t2 = tienda2["Precio"].sum()
t3 = tienda3["Precio"].sum()
t4 = tienda4["Precio"].sum()

print(f'Ingreso total de Tienda 1 --> {t1}')
print(f'Ingreso total de Tienda 2 --> {t2}')
print(f'Ingreso total de Tienda 3 --> {t3}')
print(f'Ingreso total de Tienda 4 --> {t4}')

#Creando diccionario y lista

dif_Ingresos = {'Tienda 1':t1,
               'Tienda 2':t2,
               'Tienda 3':t3,
               'Tienda 4':t4}

j = 0
for i in dif_Ingresos.values():
  if i == max(dif_Ingresos.values()):
    max_Valor = list(dif_Ingresos.keys())[j]
  elif i == min(dif_Ingresos.values()):
    min_Valor = list(dif_Ingresos.keys())[j]
  j+=1

print(f'La tienda que tiene mayores ingresos es {max_valor}')
print(f'La tienda que tiene menores ingresos es {min_valor}')
```

# 2. Ventas por categoría

```python
print('Tienda 1\n')
t1_Categorias = tienda['Categoría del Producto'].value_counts()
print(t1_Categorias,'\n')

print('Tienda 2\n')
t2_Categorias = tienda2['Categoría del Producto'].value_counts()
print(t2_Categorias,'\n')

print('Tienda 3\n')
t3_Categorias = tienda3['Categoría del Producto'].value_counts()
print(t3_Categorias,'\n')

print('Tienda 4\n')
t4_Categorias = tienda4['Categoría del Producto'].value_counts()
print(t4_Categorias,'\n')
```

# 3. Calificación promedio de la tienda


```python
p1 = tienda['Calificación'].mean()
p2 = tienda2['Calificación'].mean()
p3 = tienda3['Calificación'].mean()
p4 = tienda4['Calificación'].mean()

prom_Calificaciones = {
    'Tienda 1': p1,
    'Tienda 2': p2,
    'Tienda 3': p3,
    'Tienda 4': p4
}

#Print obtenido chatgpt "print(prom_Calificaciones.items())" print anterior"
print("⭐ Calificación Promedio por Tienda ⭐\n")
for tienda, calificacion in prom_Calificaciones.items():
    print(f"{tienda}: {calificacion:.2f}")


#Este for desempaqueta cada par (clave, valor) en dos variables:

#tienda = la clave (ej: 'Tienda 1')

#calificacion = el valor (ej: 4.46)

#Por eso, en cada vuelta del bucle puedes usar directamente esas variables.

```

# 4. Productos más y menos vendidos

```python
print('Tienda 1\n')
t1_productosmax = tienda['Producto'].value_counts().head(5)
t1_productosmin = tienda['Producto'].value_counts().tail(5)
# Concatenar ambos resultados
t1_productos = pd.concat([t1_productosmax, t1_productosmin])
print(t1_productos)

print('Tienda 2\n')
t2_productosmax = tienda2['Producto'].value_counts().head(5)
t2_productosmin = tienda2['Producto'].value_counts().tail(5)
# Concatenar ambos resultados
t2_productos = pd.concat([t2_productosmax, t2_productosmin])
print(t2_productos)

print('Tienda 3\n')
t3_productosmax = tienda3['Producto'].value_counts().head(5)
t3_productosmin = tienda3['Producto'].value_counts().tail(5)
# Concatenar ambos resultados
t3_productos = pd.concat([t3_productosmax, t3_productosmin])
print(t3_productos)

print('Tienda 4\n')
t4_productosmax = tienda4['Producto'].value_counts().head(5)
t4_productosmin = tienda4['Producto'].value_counts().tail(5)
# Concatenar ambos resultados
t4_productos = pd.concat([t4_productosmax, t4_productosmin])
print(t4_productos)
```

```python

# 2) Agrupar por tienda y sumar precios
ingresos = df.groupby("Tienda")["Precio"].sum()

# 3) Dibujar gráfico de barras
plt.figure(figsize=(8,5))
plt.bar(ingresos.index, ingresos.values)
plt.title("Ingreso total por tienda")
plt.ylabel("Ingreso total (suma de precios)")
plt.xlabel("Tienda")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Heatmap simplificado usando kdeplot

import seaborn as sns
import matplotlib.pyplot as plt

# Concatenar todos los DataFrames en uno solo
df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x='lon', y='lat', weights=df['Precio'], fill=True, cmap='Reds', bw_adjust=0.5)
plt.title('Mapa de calor de ventas (todas las tiendas)')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.grid(True)
plt.show()


```
